```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/content/daily_data.csv')
df
```

```python
df.describe()
```

```python
#TIME SERIES PLOT FOR THE WHOLE DATA SET
Col_DP = df[['DATE', 'DailyPrecipitation']]
Col_DP['DATE'] = pd.to_datetime(Col_DP['DATE'])
plt.figure(figsize=(12, 6))
plt.plot(Col_DP['DATE'], Col_DP['DailyPrecipitation'], color='b')
plt.xlabel('Date')
plt.ylabel('Value')
plt.title('Time Series Plot of Precipitation')
plt.grid(True)
plt.show()
```

```python
#TIME SERIES PLOT PER YEAR
Col_DP['DATE'] = pd.to_datetime(Col_DP['DATE'])
grouped = Col_DP.groupby(Col_DP['DATE'].dt.year)

for year, group in grouped:
    plt.figure(figsize=(12, 6))
    plt.plot(group['DATE'], group['DailyPrecipitation'], color='b')
    plt.xlabel('Date')
    plt.ylabel('Value')
    plt.title(f'Time Series Plot of Precipitation for Year {year}')
    plt.grid(True)
    plt.show()
```

```python
#MEAN
DailyPrecipitation = "DailyPrecipitation"
df[DailyPrecipitation] = pd.to_numeric(df[DailyPrecipitation], errors='coerce')
mean_value = df[DailyPrecipitation].mean()

print(f"Mean of {DailyPrecipitation}: {mean_value}")
```

```python
#MEDIAN
median_value = df[DailyPrecipitation].median()

print(f"Median of {DailyPrecipitation}: {median_value}")
```

```python
#STANDARD DEVIATION
std_deviation_value = df[DailyPrecipitation].std()


print(f"Standard Deviation of {DailyPrecipitation}: {std_deviation_value}")
```
